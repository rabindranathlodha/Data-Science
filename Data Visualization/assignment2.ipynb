```python
players = ["Virat Kohli", "KL Rahul", "Shubman Gill", "Rohit Sharma", "Shikhar Dhawan"]
runs = [3886, 1759, 2286, 1928, 2081]
```

```python
import matplotlib.pyplot as plt
width = 0.35
plt.bar(players, runs, width, color="skyblue")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center')
 
plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

```python
plt.barh(players, runs, color="red", height=0.4)
plt.xlabel("Total Runs in First 5 Years")
plt.xticks(rotation=45)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```
