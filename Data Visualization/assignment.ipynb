```python
import matplotlib.pyplot as plt
```

```python
year = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
```

```python
kohli = [1054, 623, 739, 1460, 1202, 1377, 431, 129, 302, 1377, 58]
```

```python
rahul = [4, 252, 539, 633, 468, 110, 0, 461, 137, 143, 493]
```

```python
rohit = [578, 815, 564, 1293, 1030, 1490, 171, 90, 249, 383, 157]
```

```python
plt.plot(year, kohli, color ='red', linestyle='--', label="Virat Kohli")
plt.plot(year, rahul, color ='green', linestyle='-.', label="KL Rahul")
plt.plot(year, rohit, color = 'blue',  label="Rohit Sharma")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.style.use("ggplot")
```
