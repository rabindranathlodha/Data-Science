```python
import matplotlib.pyplot as plt
names = ["Android", "iOS", "Others"]
share = [72.7, 26.9, 0.4]
```

```python
plt.title("market share of mobile OS ")
plt.pie(share, labels=names, autopct='%1.1f%%', startangle=140)
```

```python
plt.barh(names, share, color=['blue', 'green', 'red'])
plt.xlabel("Market Share (%)")
plt.ylabel("Mobile OS")
plt.grid(axis='x')
plt.tight_layout()
plt.title("Mobile Operating System Market Share", fontsize=16)
plt.show()
```
