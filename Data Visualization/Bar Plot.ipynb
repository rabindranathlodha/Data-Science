```python
import matplotlib.pyplot as plt
 
year = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
kohli = [1054, 623, 739, 1460, 1202, 1377, 431, 129, 302, 1377, 58]
```

```python
plt.bar(year, kohli)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Virat Kohli Yearly Runs")
plt.show()
```

```python
import numpy as np
 
rohit = [578, 815, 564, 1293, 1030, 1490, 171, 90, 249, 383, 157]
rahul = [4, 252, 539, 633, 468, 110, 0, 461, 137, 143, 493]
kohli = [1054, 623, 739, 1460, 1202, 1377, 431, 129, 302, 1377, 58]
 
x = np.arange(len(year))  # index positions
width = 0.25
```

```python
x
```

```python
plt.bar(x - width, rohit, width=width, label="rohit")
plt.bar(x, rahul, width=width, label="rahul")
plt.bar(x + width, kohli, width=width, label="Kohli")
 
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, year)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
```

```python
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```
